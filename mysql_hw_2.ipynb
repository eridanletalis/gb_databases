2. Для данных таблицы “sales” укажите тип заказа в зависимости от кол-ва : меньше 100 - "Маленький заказ"; от 100 до 300 - "Средний заказ"; больше 300 - "Большой заказ";
 


<code>
  SELECT id,<br />
		 &nbsp;&nbsp;&nbsp;&nbsp;CASE<br />
